```python
def make_top_url(bounds, request_id):
    part_1 = 'https://www.zillow.com/search/GetSearchPageState.htm?searchQueryState={"pagination":{},"mapBounds":{'
    part_2 = '"west":{west},"east":{east},"south":{south},"north":{north}'.format(**bounds)
    part_3 = '},"isMapVisible":true,"filterState":{"isPreMarketForeclosure":{"value":false},"isForSaleForeclosure":{"value":false},"isMultiFamily":{"value":false},"isAllHomes":{"value":true},"isAuction":{"value":false},"isNewConstruction":{"value":false},"isForRent":{"value":true},"isLotLand":{"value":false},"isManufactured":{"value":false},"isForSaleByOwner":{"value":false},"isComingSoon":{"value":false},"isPreMarketPreForeclosure":{"value":false},"isForSaleByAgent":{"value":false}},"isListVisible":true,"mapZoom":9}&wants={"cat1":["mapResults"]}&requestId='
    
    return part_1 + part_2 + part_3 + str(request_id)


def make_raw_data(zpid):
    raw_data = {"operationName":"ForRentDoubleScrollFullRenderQuery","variables":{"zpid":zpid,"contactFormRenderParameter":{"zpid":zpid,"platform":"desktop","isDoubleScroll":True}},"clientVersion":"home-details/6.0.11.0.0.hotfix-11-23-2020.d69fab8","queryId":"172653a956cf8fb0836010e71b7385e9"}
    
    return raw_data


def make_low_url(zpid):
    low_url_template = 'https://www.zillow.com/graphql/?zpid={}&contactFormRenderParameter=&queryId=172653a956cf8fb0836010e71b7385e9&operationName=ForRentDoubleScrollFullRenderQuery'

    return low_url_template.format(zpid)
```

```python
def transform_data(raw_details, zpid):
    # function transforms json details to data frame format

    # high level details
    high_level_raw = raw_details.json()['data']['property']

    high_level_dict = {'zpid': [zpid]}
    hig_level_props = ['city', 'homeStatus', 'price', 'streetAddress', 'datePosted', 'zipcode', 'latitude', 'longitude']
    for k in hig_level_props:
        high_level_dict[k] = [high_level_raw[k]]

    df_h = pd.DataFrame.from_dict(high_level_dict, orient='columns', dtype=None, columns=None)



    # low level details
    low_level_raw = raw_details.json()['data']['property']['resoFacts']

    low_level_dict = {'zpid': [zpid]}
    for k in low_level_raw.keys():
        value = low_level_raw[k]

        if type(value) != dict and k not in ['otherFacts', 'atAGlanceFacts']:
            if type(value) == list:
                try:
                    value = ' | '.join(value)
                except:
                    value = None
                    continue

            low_level_dict[k] = [value]

    df_l = pd.DataFrame.from_dict(low_level_dict, orient='columns', dtype=None, columns=None)
    
    return df_h.merge(df_l, on='zpid')
```

```python
df_all = pd.DataFrame()


broken_segments = []
broken_objects = []

for segment in bounds.keys():
    url_top = make_top_url(bounds[segment], segment)
    response_top = requests.get(url_top, headers=headers_top)

    if response_top.status_code != 200:
        time.sleep(60)
        response_top = requests.get(url_top, headers=headers_top)

    try:
        if response_top.status_code != 200:
            broken_segments.append(segment)

        else:
            
            if 'totalResultCount' in response_top.json()['cat1']['searchList'].keys():
                total_results = response_top.json()['cat1']['searchList']['totalResultCount']
            else:
                total_results = 'unknown'
                
            print('Segment: {}, Total objects: {}'.format(segment, total_results))
            for result in response_top.json()['cat1']['searchResults']['mapResults']:
                if 'zpid' in result.keys():
                    zpid = result['zpid']
                    

                    url_low = make_low_url(zpid)
                    raw_data = make_raw_data(zpid)
                    raw_details = requests.post(url_low, headers=headers_low, json=raw_data)

                    if raw_details.status_code != 200:
                        time.sleep(60)
                        raw_details = requests.post(url_low, headers=headers_low, json=raw_data)

                    try:
                        if raw_details.status_code != 200:
                            broken_objects.append(zpid)

                        else:
                            df_part = transform_data(raw_details, zpid)
                            df_all = df_all.append(df_part)

                    except:
                        continue

                    time.sleep(2)

            print('Cumulative data size: {}'.format(len(df_all)))

    except:
        continue


df_all.to_csv('zillow_rent_new_york.csv')
```
